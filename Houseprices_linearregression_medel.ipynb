```python
#Given a dataset of house prices and features like size, number of rooms, etc
```

```python
#random data
```

```python
import pandas as pd
import numpy as np

# Generate a larger dataset
np.random.seed(42)

size = np.random.randint(1000, 5000, 100)   # House sizes between 1000 and 5000 sq ft
rooms = np.random.randint(1, 8, 100)        # Number of rooms between 1 and 7
price = size * 150 + rooms * 25000 + np.random.randint(-50000, 50000, 100)  # Price based on size and rooms

data = {
    'Size': size,
    'Rooms': rooms,
    'Price': price
}
```

```python
#create a dataframe
```

```python
import pandas as pd
df=pd.DataFrame(data)
print(df)
```

```python
#define target
```

```python
#continues variable, target variable= Price(y)
#price can be dependent on both size and the room number
#analysing the data set, size is a better option
#defining the varibles
```

```python
y=df['Price']
X=df[['Size','Rooms']]
```

```python
y.shape
```

```python
X.shape
```

```python
#import train_test_split

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=2529)
```

```python
X_train
```

```python
X_test
```

```python
#importing the model selected
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
#fitting the model
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#linear equation would be y=1123.435+151.954â‹…x1+23600.123â‹…x2

```

```python
#predicting
y_pred=model.predict(X_test)
y_pred
```

```python
#testing the accuracy
from sklearn.metrics import mean_absolute_percentage_error
```

```python
error=mean_absolute_percentage_error(y_test,y_pred)
error
```

```python
accuracy = 100 - error
accuracy
```
